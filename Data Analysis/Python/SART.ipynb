```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import os

# List of CSV files
csv_files = ['SARTT1_Angle0.csv', 'SARTTask01_Angle45.csv', 'SARTTask01_Angle90.csv']

# List of columns to visualize and test for normality
columns_of_interest = ['Instability', 'Complexity', 'Variability', 'Arousal', 'Concentration', 'Division', 'Mental Capacity', 'Information Quantity', 'Familiarity']

for file in csv_files:
    # Read the CSV file
    data = pd.read_csv(file)
    
    # Perform normality tests and create histograms and Q-Q plots for each column
    fig, axes = plt.subplots(2, len(columns_of_interest), figsize=(20, 8))
    
    for i, column in enumerate(columns_of_interest):
        # Plot histograms and Q-Q plots as before
        sns.histplot(data[column].dropna(), kde=True, color='skyblue', ax=axes[0, i])
        axes[0, i].set_title(f'Histogram of {column}')
        axes[0, i].set_xlabel(column)
        axes[0, i].set_ylabel('Frequency')
        
        stats.probplot(data[column].dropna(), dist="norm", plot=axes[1, i])
        axes[1, i].set_title(f'Q-Q Plot of {column}')
        axes[1, i].set_xlabel('Theoretical quantiles')
        axes[1, i].set_ylabel('Ordered Values')
        
        # Perform Shapiro-Wilk test for normality
        stat, p = stats.shapiro(data[column].dropna())
        axes[0, i].text(0.05, 0.9, f'Shapiro-Wilk p = {p:.3f}', transform=axes[0, i].transAxes, fontsize=10, verticalalignment='top')
        
        # Determine if data is normalized based on p-value
        if p > 0.05:
            axes[0, i].text(0.05, 0.8, 'Data is normalized', transform=axes[0, i].transAxes, fontsize=10, verticalalignment='top', color='green')
        else:
            axes[0, i].text(0.05, 0.8, 'Data is not normalized', transform=axes[0, i].transAxes, fontsize=10, verticalalignment='top', color='red')
    
    fig.suptitle(f'Data Visualization and Normality Tests - {os.path.basename(file)}', fontsize=16)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import friedmanchisquare
import scikit_posthocs as sp

# List of CSV files
csv_files = ['SARTT1_Angle0.csv', 'SARTTask01_Angle45.csv', 'SARTTask01_Angle90.csv']

# List of columns to visualize and test for normality
columns_of_interest = ['Instability', 'Complexity', 'Variability', 'Arousal', 'Concentration', 'Division', 'Mental Capacity', 'Information Quantity', 'Familiarity']

# Initialize an empty list to store all data
all_data = []

# Read each CSV file into a DataFrame and store in all_data
for file in csv_files:
    data = pd.read_csv(file)
    # Convert all columns of interest to numeric data
    data[columns_of_interest] = data[columns_of_interest].apply(pd.to_numeric, errors='coerce')
    all_data.append(data)

# Combine all data into a single DataFrame
combined_data = pd.concat(all_data, keys=['Angle0', 'Angle45', 'Angle90'])

# Perform Friedman test and plot boxplots for each column
for column in columns_of_interest:
    # Extract data for the current column from each angle
    data_angle0 = combined_data.loc['Angle0'][column].dropna()
    data_angle45 = combined_data.loc['Angle45'][column].dropna()
    data_angle90 = combined_data.loc['Angle90'][column].dropna()
    
    # Perform Friedman test
    statistic, p_value = friedmanchisquare(data_angle0, data_angle45, data_angle90)
    
    # Print Friedman test results
    print(f"{column}: Friedman Test - Statistic: {statistic:.3f}, p-value: {p_value:.6f}")
    
    # Plot boxplots
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='level_0', y=column, data=combined_data.reset_index(), palette='Set3')
    plt.title(f'Boxplot of {column} by Angle')
    plt.xlabel('Angle')
    plt.ylabel(column)
    plt.show()
    
    # If p-value < 0.05, perform post hoc tests
    if p_value < 0.05:
        print(f"Post hoc tests for {column} (p < 0.05):")
        
        # Combine the data for the Nemenyi test
        data_for_posthoc = pd.DataFrame({
            'Angle0': data_angle0.values,
            'Angle45': data_angle45.values,
            'Angle90': data_angle90.values
        })
        
        # Perform Nemenyi test
        nemenyi_results = sp.posthoc_nemenyi_friedman(data_for_posthoc)
        print(f"Nemenyi test p-values:\n{nemenyi_results}\n")
        
        # Interpret Nemenyi results
        for i, (row, column) in enumerate([(0, 1), (0, 2), (1, 2)]):
            if nemenyi_results.iat[row, column] < 0.05:
                print(f"The difference between Angle{row * 45} and Angle{(column + 1) * 45} is significant (p < 0.05).")
            else:
                print(f"The difference between Angle{row * 45} and Angle{(column + 1) * 45} is not significant (p >= 0.05).")
                
        print("\n")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import friedmanchisquare

# Assuming combined_data is a DataFrame with hierarchical index where 'Angle0', 'Angle45', 'Angle90' are levels
# and columns_of_interest is a list of the columns you are interested in.

# Initialize lists to store mean values, standard deviations, and p-values
means = []
stds = []
p_values = []

# Calculate means, standard deviations, and store p-values for each column
for column in columns_of_interest:
    data_angle0 = combined_data.loc['Angle0'][column].dropna()
    data_angle45 = combined_data.loc['Angle45'][column].dropna()
    data_angle90 = combined_data.loc['Angle90'][column].dropna()
    
    # Find the minimum length of the three data sets
    min_length = min(len(data_angle0), len(data_angle45), len(data_angle90))
    
    # Truncate each data set to the minimum length
    data_angle0 = data_angle0.iloc[:min_length]
    data_angle45 = data_angle45.iloc[:min_length]
    data_angle90 = data_angle90.iloc[:min_length]
    
    # Calculate means and standard deviations
    mean_angle0 = data_angle0.mean()
    mean_angle45 = data_angle45.mean()
    mean_angle90 = data_angle90.mean()
    
    std_angle0 = data_angle0.std()
    std_angle45 = data_angle45.std()
    std_angle90 = data_angle90.std()
    
    means.append([mean_angle0, mean_angle45, mean_angle90])
    stds.append([std_angle0, std_angle45, std_angle90])
    
    # Perform Friedman test
    statistic, p_value = friedmanchisquare(data_angle0, data_angle45, data_angle90)
    p_values.append(p_value)

# Create a single bar graph
bar_width = 0.25
x = np.arange(len(columns_of_interest))

fig, ax = plt.subplots(figsize=(15, 8))

# Plot bars with error bars for standard deviations
bars0 = ax.bar(x - bar_width, [m[0] for m in means], bar_width, yerr=[s[0] for s in stds], capsize=5, label='Angle0')
bars45 = ax.bar(x, [m[1] for m in means], bar_width, yerr=[s[1] for s in stds], capsize=5, label='Angle45')
bars90 = ax.bar(x + bar_width, [m[2] for m in means], bar_width, yerr=[s[2] for s in stds], capsize=5, label='Angle90')

# Define a text offset for the p-value labels
text_offset = 0.2  # Adjust this value to change the y position of the p-value text

# Determine the current y-axis limits
y_min, y_max = ax.get_ylim()

# Annotate p-values and add asterisks for significant differences
for i in range(len(columns_of_interest)):
    if p_values[i] < 0.05:
        annotation_y = min(max(means[i]) + max(stds[i]) + text_offset, y_max - text_offset)
        ax.text(x[i], annotation_y, f'*p={p_values[i]:.4f}', ha='center', va='bottom', color='green')

# Add labels and title
ax.set_xlabel('SART Metrics')
ax.set_ylabel('Mean Values')
ax.set_title('Mean SART Metrics by Angle with Standard Deviations and p-values')
ax.set_xticks(x)
ax.set_xticklabels(columns_of_interest)
ax.legend()

plt.tight_layout()
plt.savefig('sart_task01.png')
plt.show()
```
